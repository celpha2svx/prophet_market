# Data Visualisation

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the features
df = pd.read_csv('aapl_features.csv')
df['date'] = pd.to_datetime(df['date'])

# Filter to last 6 months for clarity
df_recent = df.tail(180)  # Approximately 6 months of trading days

# Create figure with subplots
fig, axes = plt.subplots(3, 1, figsize=(15, 12))

# ============ PLOT 1: Price + Moving Averages ============
ax1 = axes[0]
ax1.plot(df_recent['date'], df_recent['close'], label='Close Price', color='black', linewidth=2)
ax1.plot(df_recent['date'], df_recent['sma_20'], label='SMA 20', color='blue', linestyle='--')
ax1.plot(df_recent['date'], df_recent['sma_50'], label='SMA 50', color='red', linestyle='--')

# Mark Golden Cross (SMA20 crosses above SMA50)
sma20_above = df_recent['sma_20'] > df_recent['sma_50']
crossovers = sma20_above != sma20_above.shift()
golden_crosses = df_recent[crossovers & sma20_above]
death_crosses = df_recent[crossovers & ~sma20_above]

ax1.scatter(golden_crosses['date'], golden_crosses['close'], 
           color='green', marker='^', s=200, label='Golden Cross', zorder=5)
ax1.scatter(death_crosses['date'], death_crosses['close'], 
           color='red', marker='v', s=200, label='Death Cross', zorder=5)

ax1.set_title('AAPL Price with Moving Averages', fontsize=16, fontweight='bold')
ax1.set_ylabel('Price ($)', fontsize=12)
ax1.legend(loc='best')
ax1.grid(True, alpha=0.3)

# ============ PLOT 2: RSI ============
ax2 = axes[1]
ax2.plot(df_recent['date'], df_recent['rsi'], label='RSI', color='purple', linewidth=2)
ax2.axhline(y=70, color='r', linestyle='--', label='Overbought (70)')
ax2.axhline(y=30, color='g', linestyle='--', label='Oversold (30)')
ax2.fill_between(df_recent['date'], 70, 100, alpha=0.2, color='red')
ax2.fill_between(df_recent['date'], 0, 30, alpha=0.2, color='green')

ax2.set_title('Relative Strength Index (RSI)', fontsize=16, fontweight='bold')
ax2.set_ylabel('RSI', fontsize=12)
ax2.set_ylim(0, 100)
ax2.legend(loc='best')
ax2.grid(True, alpha=0.3)

# ============ PLOT 3: MACD ============
ax3 = axes[2]
ax3.plot(df_recent['date'], df_recent['macd'], label='MACD', color='blue', linewidth=2)
ax3.plot(df_recent['date'], df_recent['macd_signal'], label='Signal Line', color='red', linewidth=2)
ax3.bar(df_recent['date'], df_recent['macd_histogram'], label='Histogram', 
        color=['green' if x > 0 else 'red' for x in df_recent['macd_histogram']], alpha=0.5)
ax3.axhline(y=0, color='black', linestyle='-', linewidth=0.5)

ax3.set_title('MACD (Moving Average Convergence Divergence)', fontsize=16, fontweight='bold')
ax3.set_xlabel('Date', fontsize=12)
ax3.set_ylabel('MACD', fontsize=12)
ax3.legend(loc='best')
ax3.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('aapl_technical_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n Chart saved as 'aapl_technical_analysis.png'")
```
